```r
# Set queue model input parameters
input_mm1 <- NewInput.MM1(lambda = 1/6, mu = 1/2, n = 100)

# Create queue class object
output_mm1 <- QueueingModel(input_mm1)

# Get queue model report
Report(output_mm1)

# Get queue model summary
summary(output_mm1)
```

```r
#bisa ditambah set.seed()
arrivals <- cumsum(rexp(25, 1/6))
print(arrivals)
service <- rexp(25,1/2)

queue_mm1 <- queue_step(arrivals = arrivals, service = service)

#str(queue_mm1,1)
summary(queue_mm1)
```
